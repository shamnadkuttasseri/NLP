### Stemming

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

sentence=nltk.sent_tokenize(paragraph)
stemmer=PorterStemmer()

#stemming
#this code segment applies stemming to each word in the sentences extracted from the paragraph while filtering out stopwords,
#resulting in a list of stemmed sentences

for i in range (len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)
print(sentences)
```

### Lemetization

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

sentences=nltk.sent_tokenize(paragraph)
lemmatizer=WordNetLemmatizer()

#Lemmatization
#This loop iterates over each sentence in the list of sentences in the list of sentances,tokenizes the current sentence into 
#words
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    #lemmatize each word  in the sentence using the WordNet lemmatization.it also removes stopwords from the sentence
    sentences[i]=' '.join(words)
    #joins the lemmatized   words back into a sentence and assigns it to corresponding index in the list of seneteces
    print(sentences)
```

### bag of words

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

ps=PorterStemmer()
wordnet=WordNetLemmatizer()
sentences=nltk.sent_tokenize(paragraph)
corpus=[]
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z]',' ',sentences[i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word)for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
    
#     loop through each sentence:
#     Remove non alphabeeetic characters using regular expressions.
#     convert the sentence to lowercase.
#     Split the sentence into words
#     stem each word using the PorterStemmer(ps)
#     join the stemmed words back into sentence.
#     Build corpus:Append the preprocessed sentences to the corpus list.
```

### Chunking

sequence of words(usually a sentence) is segmented into syntically correlated groups called 'chunks'

```python
import nltk

chunker=nltk.RegexpParser(r'''
    NP:{<DT><NN.*><.*><NN.*>}
    }<VB.*>{
    '''
)

#tokenize the paragraph
tokens=nltk.word_tokenize(paragraph)

#perform part of speech tagging
tagged=nltk.pos_tag(tokens)

#parse the sentence
parsed_output=chunker.parse(tagged)

#display the parsed output
print(parsed_output)

# nltk.RegexpParser: This is a tool from NLTK (Natural Language Toolkit) that helps us create a chunk parser using regular expressions.

# r'''...''': We use triple quotes to create a raw string literal, 
# which means that backslashes are treated as normal characters and don't need to be escaped.
# NP:{<DT><NN.*><.*>*<NN.*>}: This is a pattern that defines a noun phrase (NP). It says that an NP starts with a determiner 
# (like "the" or "a", indicated by <DT>), followed by one or more nouns (like "cat" or "dog", indicated by <NN*>),
# then followed by zero or more of any token (indicated by <.*>*), and finally ending with another noun.

# }<VB.*>{: This part indicates that a chunk (in this case, a noun phrase) should end if it encounters a verb
# The notation <VB*> represents any form of
# The notation <VB*> represents any form of a verb.
```

### NER (Named Entity Recognition)

```python
from nltk import ne_chunk,pos_tag,word_tokenize
```

```python
def get_near(text):
    i=ne_chunk(pos_tag(word_tokenize(text)),binary=True)
    return[a for a in i if len(a)==1]
get_near(paragraph)

#word_tokenize(text) tokenizes the input text into individual words.
#pos_tag() assigns POS tags to each word
#ne_chunk performs named entity chunking on the tagged words
# the binary=True parameter specifics that only named entities are chunked,
# and non-entities are ignored.the result is stored in the variable i
# i conatains the results of named entity chunking performed in the input text.
# each named entities chunk is represented as a tree-like structure (hea tree object)
# , and entities are nested within it.
# in the context of part of speech tagging (POS tagging),"NNP" stands for "Proper Noun,singular"
```
